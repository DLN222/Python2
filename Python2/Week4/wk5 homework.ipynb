```python
urls_lc = [f"{url}-{url_number}" for url_number in range(2,11)]
urls_lc
```

```python
from random import randrange
import time
```

```python

artists = []
for url in urls_lc:
    try: 
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        artist = soup.find("div", class_= "artist")
        for art in artist:
            artists.append(artist.get_text())
    except:
        print(f"There's a problem with...{url}")
    finally:
        snoozer = randrange(2,5)
        print(f"Snoozing for {snoozer} before next")
        time.sleep(snoozer)
```
