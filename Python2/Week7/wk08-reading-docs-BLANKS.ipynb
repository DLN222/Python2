````python
import glob
import pandas as pd
````

````python
target_files = sorted(glob.glob("project-docs/*txt"))
target_files
````

````python

for target_file in target_files:
    with open(target_file, "r") as my_text:
        all_text = my_text.readlines()
        print(f"Source file: {target_file}")
        print(f"{all_text[0].rstrip()}")
        medical_condition = print(f"Medical Condition: {all_text[2].rstrip()}")
        print(f"Care Hours: {all_text[4]}")
        print(f"")
      
````

````python
target_list = []

for target_file in target_files:
    with open(target_file, "r") as my_text:
        all_text = my_text.readlines()
        print(f"Source file: {target_file}")
        year = all_text[0].replace("Year: ", "").rstrip()
        print(year)
        condition = all_text[2].rstrip()
        if "Dementia" in condition:
            condition = "Dementia"
        elif "Alzheimer's" in condition:
            condition = "Alzheimer's"
        else:
            condition = "Not Specified"
        print(condition)
        care = all_text[4]
        if "24-hour" in care:
            care = "Full Day"
        elif "partial" in care:
            care = "Half Day"
        else:
            care = "Not Specified"
        print(care)
        if "Dementia" in condition:
            cognition_related = "TRUE"
        else:
            cognition_related = "FALSE"
        print(f"")
        
        
        target_list.append({"Year":year, 
                            "Medical Condition":condition, 
                            "Care Hours": care,
                           "Cognition Related": cognition_related})
        
````

````python
df = pd.DataFrame(target_list)
df
````
