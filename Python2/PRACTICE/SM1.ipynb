```python
import pandas as pd
import glob
import pymupdf4llm
from mangoCR import pdf2image_ocr
import os
import fitz
```

```python
pdf_files = glob.glob("Originals-Court Documents/*.pdf")
pdf_files
```

```python
file_names = []
page_counts = []

for file in pdf_files:
    names = file.replace("Originals-Court Documents/", "")
    file_names.append(names)
    
    with fitz.open(file) as pdf_document:
        num_pages = pdf_document.page_count
        page_counts.append(num_pages)
        
df = pd.DataFrame({
    "File Names": file_names,
    "Page_Count": page_counts
})

df
```

```python
pdf_files = 'Originals-Court Documents/*.pdf'
pdf2image_ocr(pdf_files,"{file_name}.md")
```

```python
mangos = []  # To store markdown filenames

for file in pdf_files:
    # Extract text using OCR
    p2i = pdf2image_ocr(file)  # This should return a string with the extracted text

    # Debug: Check the output of the OCR function
    if p2i is None:
        print(f"Warning: OCR failed for {file}, skipping.")
        continue  # Skip to the next file if OCR fails

    # Create a markdown filename
    markdown_filename = os.path.basename(file).replace(".pdf", ".md")
    print(f"Writing to markdown file: {markdown_filename}")

    # Write the extracted text to the markdown file
    with open(markdown_filename, "w") as md_file:
        md_file.write(p2i)  # Write the text to the file

    # Add the markdown filename to the list
    mangos.append(markdown_filename)

# Print the list of markdown files created
print("Markdown files created:", mangos)
```
