## Prompt

Imagine you want to send the above analytics information (that came from somewhere like your internal Google Analytics platform) to ChatGPT to be analyzed to spot trends and forecast traffic.

1. What PII is in here that needs to be protected?
2. How could you mask that PII so that this data could be passed to ChatGPT?

*Fun fact:* ChatGPT generated most of this CSV, but it forgot to enclose the `Location` field in quotes, which threw me off for a long time.

```python
#df['id'] = df['RegistryID'].apply(lambda x: 123456789123 - x.year)
#df['id'] = df['RegistryID'].apply(lambda x: '*' * (len(str(x)) - 4) + str(x)[-4:])
mask_id = df['RegistryID'] = df['RegistryID'].apply(lambda x: '***' + str(x)[3:]) #First 3 Digits as Stars
```

```python
df.head()
```

```python
df['FacStreet'] = df['FacStreet'].apply(
    lambda x: x[0] + '*' * (len(x) - 2) + x[-1] if len(x) > 1 else '*'
)

df
```

```python
df['FacName'] = df['FacName'].apply(
    lambda x: x[0] + '*' * (len(x) - 2) + x[-1] if len(x) > 1 else '*'
)

df
```

```python
df['DFR URL'] = df['DFR URL'].apply(
    lambda x: x.split('/')[0] + '//' + x.split('/')[2] + '/****' if '//' in x else '****'
)

df
```

```python
city_to_code = {
    "NY": 0,
    "Brooklyn": 1,
    
}
```

```python
df['FacCity'] = df['FacCity'].apply(lambda x: city_to_code.get(x))
```

```python
pii_columns = ['FacName', 'FacStreet', 'FacCity']
```

```python
columns_to_use = [ c for c in df.columns if c not in pii_columns ]
```

```python
df_clean = df[columns_to_use].copy()
```
